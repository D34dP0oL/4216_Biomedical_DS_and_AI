```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

#### 1. Load the dataset from the given dataset.tsv file.

```python
df  = pd.read_csv("dataset.tsv", sep='\t')
df.drop(df.columns[0], axis=1, inplace=True)
```

#### 2. Plot the scatterplot matrix for the given dataset. What can be seen in the scatterplot matrix?

There is a high positive correlation between feature 1 & 3, there is also a positive correlation between feature 1 & 2 and also between feature 2 & 3.

```python
pd.plotting.scatter_matrix(df, alpha=1)
```

#### 3. Which correlation would suit the comparison of feature_1 and feature_3?

Spearman's rank correlation works good because we have non-linear dependencies between feature 1 & 3
#### Calculate the relevant correlation coefficient for the 2 features.

```python
correlation = df.corr(method='spearman')
print('Spearman correlation between feature 1 & 3 is', correlation.iloc[0,2])
```

#### 4. Plot the correlation heatmap of the entire dataset.

```python
sn.heatmap(correlation, square=True, linewidths=.5)
```
